```python
n=11
xb = np.linspace(0,np.pi/np.sqrt(2),n)
yb = np.linspace(0,np.pi/np.sqrt(2),n)
dx = xb[1] - xb[0]
dy = yb[1] - yb[0]
x = np.concatenate(([-3*dx/2, -dx/2], xb + (dx/2), [xb[-1] + (3*dx/2)]))
y = np.concatenate(([-3*dy/2, -dy/2], yb + (dy/2), [yb[-1] + (3*dy/2)]))
coords =  np.array([[[x_,y_] for x_ in x] for y_ in y])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
true = np.sin((coords[2:-2,2:-2,0] + coords[2:-2,2:-2,1])/np.sqrt(2))
ax.plot_wireframe(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], true,color='red',alpha=0.5)
ax.plot_wireframe(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], -true,alpha=0.5)
# true = np.sin((coords[:,:,0] + coords[:,:,1])/np.sqrt(2))
# ax.plot_wireframe(coords[:,:,0], coords[:,:,1], true,color='red')
# ax.plot_wireframe(coords[:,:,0], coords[:,:,1], -true)
# ax.contour(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], true,alpha=0.5)
# ax.contour(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], -true,alpha=0.5)
elev = 0 # Up and down angle (0 is horizontal, 90 is vertical)
azim = -60 # Left and right angle (0 is facing front)
ax.view_init(elev=elev, azim=azim)
# plt.close()
```
