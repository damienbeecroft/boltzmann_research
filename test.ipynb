```python
from numba import cfunc

@cfunc("float64(float64, float64)")
def add(x, y):
    return x + y

print(add.ctypes(4.0, 5.0))  # prints "9.0"
```

```python
from numba import njit, cfunc, types

@cfunc(types.float64(types.float64))
def my_callback(x):
    return x * 2

@njit
def sweep(callback, x):
    # Use the callback as a regular function
    return callback(x)

# Now you can pass the address of the compiled callback
result = sweep(my_callback.address, 5.0)
```
