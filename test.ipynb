```python
n=11
xb = np.linspace(0,np.pi/np.sqrt(2),n)
yb = np.linspace(0,np.pi/np.sqrt(2),n)
dx = xb[1] - xb[0]
dy = yb[1] - yb[0]
x = np.concatenate(([-3*dx/2, -dx/2], xb + (dx/2), [xb[-1] + (3*dx/2)]))
y = np.concatenate(([-3*dy/2, -dy/2], yb + (dy/2), [yb[-1] + (3*dy/2)]))
coords =  np.array([[[x_,y_] for x_ in x] for y_ in y])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
true = np.sin((coords[2:-2,2:-2,0] + coords[2:-2,2:-2,1])/np.sqrt(2))
ax.plot_wireframe(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], true,color='red',alpha=0.5)
ax.plot_wireframe(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], -true,alpha=0.5)
# true = np.sin((coords[:,:,0] + coords[:,:,1])/np.sqrt(2))
# ax.plot_wireframe(coords[:,:,0], coords[:,:,1], true,color='red')
# ax.plot_wireframe(coords[:,:,0], coords[:,:,1], -true)
# ax.contour(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], true,alpha=0.5)
# ax.contour(coords[2:-2,2:-2,0], coords[2:-2,2:-2,1], -true,alpha=0.5)
elev = 0 # Up and down angle (0 is horizontal, 90 is vertical)
azim = -60 # Left and right angle (0 is facing front)
ax.view_init(elev=elev, azim=azim)
# plt.close()
```

```python
import numpy as np

# Example parameters (replace these with your actual values)
Lx, Lv = 10, 10
Nx, Nv = 100, 100

# Defining the meshgrid
xb = np.linspace(-Lx/2, Lx/2, Nx) # spatial cell boundaries
dx = xb[1] - xb[0]
x_grid = np.concatenate(([-dx/2 + xb[0]], xb + (dx/2))) # spatial cell centers
v_grid = np.linspace(-Lv/2, Lv/2, Nv)
X, V1, V2 = np.meshgrid(x_grid, v_grid, v_grid, indexing='ij')

# Assuming f0 is defined elsewhere and computes the value of f at each point in the grid
# For example purpose, let's define a dummy f0
def f0(X, V1, V2):
    return np.exp(-V1**2 - V2**2) * np.sin(X)

f = f0(X, V1, V2)

# Integrate over V1 and V2
integral_over_V1 = np.trapz(f, x=v_grid, axis=1) # This integrates out V1
integral_over_V1_V2 = np.trapz(integral_over_V1, x=v_grid, axis=1) # This integrates out V2 as well

# Now integral_over_V1_V2 is a vector where each element corresponds to the integral of f over V1 and V2 at each X
```
